```python
import pandas as pd
import numpy as np
```

```python
input_file = "task1_test_gpt35_eval.json"
results = pd.read_json(f"../data/{input_file}")
results['num_constraints'] = results['constraints'].apply(len)
results['num_constraints_sat'] = results['positive_constraints'].apply(lambda x: np.mean([np.count_nonzero(r) for r in x]) if x else None) / results['num_constraints']
```

```python
#helpers
grand_mean = lambda responses: np.mean(responses.dropna().apply(np.mean))
```

```python
def get_aggs(results, by='num_constraints'):
    aggs = results.groupby(by).agg({"Distinctiveness": ['mean'],
                                              'num_constraints_sat': ['mean'],
                                              **{col: grand_mean for col in results.columns[9:13]}})
    return aggs
```

```python
get_aggs(results)
```

```python
get_aggs(results, ["n_subcats"])
```

```python
get_aggs(results, ["n_subcats","num_constraints"])
```
